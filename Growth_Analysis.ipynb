```python
# relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

### Loading and Exploring the PSP dataset

```python
# convert excel file to csv file
def xlx_to_csv(file_path, sheetname):
    xls = pd.read_excel(file_path, sheetname, index_col=None)
    csv_file= xls.to_csv('psp_data.csv', encoding='utf-8')
    return csv_file
```

```python
# Loading the dataset
xlx_to_csv('PSP_DATA.xlsx', 'PSP')

# Load the dataset to pandas dataframe
df = pd.read_csv('psp_data.csv')

# view a snapshot of the dataset
df.head()
```

```python
df.info()
```

### Data Cleaning and Preparation

```python
df = df[['Day', 'Month', ' Year', 'Planting Year', 'Block', 'Plot Number', 'Slope', 'Diseases', 'Fire Scars', 'Beetle', 'Tree Number', 'Height', 'DBH (cm)']]
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# create measurement date
df['Month'] = df['Month'].astype(str)
df[' Year'] = df[' Year'].astype(str)
```

```python
df.info()
```

```python
df
```

```python
df.info()
```

```python
df['Day'] = df['Day'].astype(str)
df = df[~((df['Day'] == '  '))]
```

```python
df = df.drop('Measurement date', axis=1)
```

```python
df['Measurement date'] = df['Day'] + '/' + df['Month'] + '/' + df[' Year']
```

```python
df['Day'].unique()
```

```python
df['Measurement date'] = pd.to_datetime(df['Measurement date'], dayfirst=True)
```

```python
df.info()
```

```python
df['Diseases'].unique()
```

### Remove trees with diseases

```python
df = df[~((df['Diseases'] == 'yes'))]
```

```python
df['Fire Scars'].unique()
```

### Remove trees with fire scars

```python
df = df[~((df['Fire Scars'] == 'yes'))]
```

#### Remove trees with beetle attack

```python
df['Beetle'].unique()
```

```python
df = df[~((df['Beetle'] == 'Yes') | (df['Beetle'] == 'yes'))]
```

```python
df.info()
```

### Tree Age

```python
df['Planting Year'].unique()
```

```python
## Drop UNKNOWN
df = df[~((df['Planting Year'] == 'UNKNOWN'))]
```

```python
df['age'] = (df['Measurement date'] - pd.to_datetime(df['Planting Year'])).dt.days / 365.25
```

```python
df.head()
```

### Compute for Basal Area

```python
df['DBH (cm)'].unique()
```

```python
## drop dead trees
df = df[~((df['DBH (cm)'] == 'dead'))]
```

```python
df['DBH (cm)'] = df['DBH (cm)'].astype(float)
df['Height'] = df['Height'].astype(float)
```

### Compute for Basal Area of the trees

```python
df['BA'] = np.pi * ((df['DBH (cm)']/200))**2
```

```python
df.head()
```

### Compute for Volume of trees

```python
df['Volume'] = 0.00007854 * (df['DBH (cm)'])**2 * df['Height']
```

### Growth Analysis

```python
# Individual Tree Growth Analysis
# Analyze growth of individual tree over time
tree_data = df.sort_values(by=['Plot Number', 'Tree Number', 'Measurement date'])
```

```python
# calculate periodic growth
tree_data['dbh_growth'] = tree_data['DBH (cm)'].diff()
tree_data['height_growth'] = tree_data['Height'].diff()
tree_data['volume_growth'] = tree_data['Volume'].diff()
tree_data['period'] = tree_data['Measurement date'].diff().dt.days / 365.25
```

```python
# Annual growth rates
tree_data['dbh_annual_growth'] = tree_data['dbh_growth'] / tree_data['period']
tree_data['height_annual_growth'] = tree_data['height_growth'] / tree_data['period']
tree_data['volume_annual_growth'] = tree_data['volume_growth'] / tree_data['period']
```
