```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# mount google drive to this notebook
from google.colab import drive

drive.mount('/content/drive')

# connect to the data using pandas dataframe
data = '/content/drive/My Drive/FG_DATA/2021_TAIN_II_DATA.xlsx'

df = pd.read_excel(data, 'PSP TEAK 2021 VI', index_col=None)

df.head()
```

```python
df = df[[' Year', 'Planting Year', 'Block', 'Plot Number', 'Diseases', 'Beetle', 'Tree Number', 'Height', 'DBH', 'Remarks']]
df.head()
```

```python
df = df[[' Year','Block', 'Planting Year', 'Plot Number', 'Tree Number', 'Height', 'DBH', 'Diseases', 'Beetle', 'Remarks']]
df.head()
```

```python
df['Remarks'].unique()
```

```python
df = df[~((df['Remarks'] == 'Beetle') | (df['Remarks'] == 'Dieback'))]
```

```python
df.head()
```

```python
df['Remarks'].unique()
```

```python
df.info()
```

### Compute Age of Trees

```python
df['Planting Year'].unique()
```

```python
# Replace 'UNKNOWN' with 2000
df['Planting Year'] = df['Planting Year'].replace('UNKNOWN', '2000')
```

```python
# convert data from str to float numbers
df[' Year'] = df[' Year'].astype(int)
df['Planting Year'] = df['Planting Year'].astype(int)
```

```python
df['Age'] = df[' Year'] - df['Planting Year']
df.head()
```

```python
df['Age'].unique()
```

### Compute Basal Area of individual Tree

```python
df['DBH'].unique()
```

```python
df = df[~((df['DBH'] == 'dead'))]
```

```python
df['DBH'] = df['DBH'].astype(float)
```

```python
df.info()
```

```python
df['Basal_area_m2'] = (np.pi * df['DBH']**2) / 40000
```

```python
df.head()
```

### Compute Basal Area per Hectare

```python
# Sum basal area per plot
ba_plot = df['Basal_area_m2'].sum()

# convert hectare
plot_area = 0.08 # 800sq.meter
df['basal_area_per_ha'] = ba_plot / plot_area
```

```python
df.head()
```

### Compute Volume per Tree

```python
form_factor = 0.45 # replace with the actual form factor constant
df['Volume_per_tree'] = df['Basal_area_m2'] * df['Height'] * form_factor
```

```python
df.head()
```

### Compute Volume per hectare

```python
total_volume = df['Volume_per_tree'].sum()

df['Volume_per_ha'] = total_volume / plot_area
```
